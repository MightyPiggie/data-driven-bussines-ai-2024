```python
data = pd.read_csv(r"C:\GitHub\jaar 2\blok 1\groeps project proraills\dataset-prorail-clean-2.csv")
single_or_empty_columns = {col: data[col].unique() for col in data.columns if data[col].nunique() == 1 or data[col].isna().all()}

if single_or_empty_columns:
    print("Columns with a single unique value or that are entirely empty:")
    for col, values in single_or_empty_columns.items():
        if data[col].isna().all():
            print(f"{col}: Empty column")
        else:
            print(f"{col}: Single unique value - {values[0]}")
else:
    print("No columns with a single unique value or that are entirely empty found.")
```

```python
# Ensure all columns and rows are displayed
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

# Calculate the total number of rows
total_rows = len(data)

# Create a DataFrame summarizing missing values
missing_values_summary = pd.DataFrame({
    'Column Name': data.columns,
    'Missing Values': data.isnull().sum(),
    'Total Values': total_rows
})

# Calculate the percentage of missing values and round it for readability
missing_values_summary['Percentage Missing'] = ((missing_values_summary['Missing Values'] / missing_values_summary['Total Values']) * 100).round(2)

# Filter to include only columns with missing values
missing_values_summary = missing_values_summary[missing_values_summary['Missing Values'] > 0]

# Print the summary in a more readable format
print("\nMissing Values Summary (Only Columns with Missing Values):\n")
print(missing_values_summary.to_string(index=False, col_space=20))
```

```python
top_n = 20  # Adjust this number as needed
top_incidents = data['stm_sap_meldtekst'].value_counts().nlargest(top_n)
plt.figure(figsize=(10, 6))
sns.barplot(y=top_incidents.index, x=top_incidents.values)
plt.title("Top Incident Types")
plt.xlabel("Count")
plt.ylabel("Incident Type")
plt.show()
```

```python
threshold = 100  # Set a threshold for minimum counts
counts = data['stm_sap_meldtekst'].value_counts()

# Apply the lambda function, handling NaN values
data['grouped_incident'] = data['stm_sap_meldtekst'].apply(
    lambda x: x if pd.notna(x) and counts[x] >= threshold else 'Other'
)

# Plot the grouped incident types
plt.figure(figsize=(10, 6))
sns.countplot(y='grouped_incident', data=data)
plt.title("Grouped Incident Types")
plt.xlabel("Count")
plt.ylabel("Incident Type")
plt.show()
```

```python
# Set the sample size for counting incidents
sample_size = 10  # Adjust this number

# Sample the data
sampled_data = data.sample(n=sample_size, random_state=1)

# Count incidents by type
incident_counts = sampled_data['stm_sap_meldtekst'].value_counts().reset_index()
incident_counts.columns = ['Incident Type', 'Count']

# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data=incident_counts, x='Incident Type', y='Count', palette='viridis')
plt.title("Incident Counts by Type (Sampled Data)")
plt.xticks(rotation=45)
plt.ylabel("Number of Incidents")
plt.show()
```

```python
# Set the sample size for the box plot
sample_size = 1000  # Adjust this number

# Sample the data
sampled_data = data.sample(n=sample_size, random_state=1)

# Create a box plot for one of the numeric features
plt.figure(figsize=(12, 6))
sns.boxplot(data=sampled_data, x='stm_prioriteit', y='stm_km_tot_mld')
plt.title("Box Plot of Total KM by Priority (Sampled Data)")
plt.xlabel("Priority")
plt.ylabel("Total KM")
plt.show()
```
