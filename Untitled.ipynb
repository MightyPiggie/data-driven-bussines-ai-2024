```python
data = pd.read_csv('dataset-prorail-clean.csv') 
single_or_empty_columns = {col: data[col].unique() for col in data.columns if data[col].nunique() == 1 or data[col].isna().all()}

if single_or_empty_columns:
    print("Columns with a single unique value or that are entirely empty:")
    for col, values in single_or_empty_columns.items():
        if data[col].isna().all():
            print(f"{col}: Empty column")
        else:
            print(f"{col}: Single unique value - {values[0]}")
else:
    print("No columns with a single unique value or that are entirely empty found.")
```

```python
pd.set_option('display.max_columns', None) #replace n with the number of columns you want to see completely
pd.set_option('display.max_rows', None) #replace n with the number of rows you want to see completely

total_rows = len(data)

missing_values_summary = pd.DataFrame({
    'Column Name': data.columns,
    'Missing Values': data.isnull().sum(),
    'Total Values': total_rows
})

missing_values_summary['Percentage Missing'] = (missing_values_summary['Missing Values'] / missing_values_summary['Total Values']) * 100

missing_values_summary = missing_values_summary[missing_values_summary['Missing Values'] > 0]

print("\nMissing values summary (only columns with missing values):")
print(missing_values_summary)
```
